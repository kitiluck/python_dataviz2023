### เริ่มทำ PCA

sklearn-> scikit-learn เป็น package ที่รวบรวม function การทำ Data MIning เอาไว้ให้เราใช้แบบไม่ต้องเขียนเอง

import

```python
from sklearn.decomposition import PCA # โหลด package PCA
```

define

```python
pca = PCA()
```

fit-transform

```python
thisdata = example_df.iloc[:,:-1]
thisdata
```

```python
new_axis = pca.fit_transform(thisdata)
```

```python
new_axis.shape # ดูขนาด
```

```python
new_axis # data ทั้งหมด
```

```python
PCAdf = pd.DataFrame(new_axis, columns = ['PCA1','PCA2','PCA3','PCA4'])
PCAdf # เอา new_axis มาใส่ตารางเพื่อให้ดูได้ง่ายขึ้น
```

```python
pca.explained_variance_ratio_ # ดูการกระจายของ data บวกตามสมาชิกไปเรื่อย ๆ (ถ้าบวกกันทั้งหมดจะได้ 100%)
```
