# เชื่อมตารางข้อมูลโดยใช้ merge

```python
dt_merge = info_province.merge(result_constituencies_status, how ='left', left_on='prov_id',right_on='prov_id')
print(dt_merge.shape)
dt_merge.head()
```

# เช็คค่า NULL

```python
print(dt_merge5.isnull().any())
dt_merge5.head()
#.isnull() เพื่อระบุว่านิพจน์มีค่า Null หรือไม่
#.any() เพื่อดูรายละเอียดการเปรียบเทียบข้อมูลให้ออกมาในรูป true false
# Ture = มีค่า Null
```

  จากการพิจารณาเพื่อเช็คค่า NULL ในแต่ละ Column โดยพิจารณาจากการอ่านค่า True และ False ซึ่งถ้ามี True = มีค่า NULL
    
  ดังนั้นจะเห็นว่า column ในตารางที่มีค่า NULL ก็คือ 4 region, party_no, zone  ซึ่งเป็นข้อมูลที่ต้องการใช้ และเสียข้อมูลเป็นจำนวนมากเกินไป
  
  จึงไม่เหมาะกับการตัดข้อมูลโดยใช้ .dropna จึงพิจารณาใช้วิธีแทนค่าใน Column ที่เราต้องการใช้งาน

# กำหนดค่าเพื่อแทนค่าใน NULL ของ Column ที่ต้องการใช้งาน

```python
import numpy as np
```

```python
# กำหนดค่า แทน Null ใน party_no
# party_no = เลขพรรคการเมืองในบัตรเลือกตั้ง
# พลังบูรพา - 97
# รักษ์ธรรม - 98
# ประชาสามัคคี - 99

dt_merge5['party_no'] = np.where(dt_merge5['name'] == 'พลังบูรพา', '97', dt_merge5['party_no'])
dt_merge5['party_no'] = np.where(dt_merge5['name'] == 'รักษ์ธรรม', '98', dt_merge5['party_no'])
dt_merge5['party_no'] = np.where(dt_merge5['name'] == 'ประชาสามัคคี', '99', dt_merge5['party_no'])
```

```python
# กำหนดค่า แทน Null ใน 4 region
# 4 region = 4 ภูมิภาค
# เชียงราย - เหนือ

dt_merge5['4 region'] = np.where(dt_merge5['province'] == 'เชียงราย', 'เหนือ', dt_merge5['4 region'])
```

# เช็คค่า NULL อีกรอบว่า ข้อมูลไม่มีค่า NULL แล้ว

```python
print(dt_merge5.isnull().any())
dt_merge5.head()
```

จากการพิจารณาค่า NULL พบว่าข้อมูลใน Column ไม่มีค่า NULL แล้ว
